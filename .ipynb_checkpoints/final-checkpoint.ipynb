```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from binance.client import Client

# Binance APIAPI
client = Client()

# 
symbol = "BTCUSDT"  # BinanceBTC/USDBTC/USDT
interval = Client.KLINE_INTERVAL_1HOUR  # 1

# 3
start_date = "2021-01-31"
end_date = "2024-01-31"

# Binance API
klines = client.get_historical_klines(symbol, interval, start_date, end_date)

# DataFrame
df = pd.DataFrame(klines, columns=[
    "timestamp", "Open", "High", "Low", "Close", "Volume",
    "Close_time", "Quote_asset_volume", "Number_of_trades",
    "Taker_buy_base_asset_volume", "Taker_buy_quote_asset_volume", "Ignore"
])

# 
df = df[["timestamp", "Open", "High", "Low", "Close", "Volume"]]

# 
df["timestamp"] = pd.to_datetime(df["timestamp"], unit="ms")  # 
df[["Open", "High", "Low", "Close", "Volume"]] = df[["Open", "High", "Low", "Close", "Volume"]].astype(float)

# 
df.set_index("timestamp", inplace=True)

# 
print(df.head())  # 5
print(df.tail())  # 5

# 
plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Close"], label="BTC/USD Close Price")
plt.xlabel("Date")
plt.ylabel("Price (USDT)")
plt.title("BTC/USD 1-hour Price Chart (Last 3 years)")
plt.legend()
plt.show()
```

```python
import numpy as np
import seaborn as sns

# 1 
print(" :")
print(df.isnull().sum())

# 
df.fillna(method="ffill", inplace=True)

# 2 
# 
df["return"] = df["Close"].pct_change()

# 5% 
outliers = df[(df["return"] > 0.05) | (df["return"] < -0.05)]

print(f" : {len(outliers)}")
print(outliers)

# 3 
df = df[(df["return"] <= 0.05) & (df["return"] >= -0.05)]

# 4 
sns.boxplot(x=df["return"])
plt.title("BTC/USD Hourly Return Distribution")
plt.show()

# 5 return 
df.drop(columns=["return"], inplace=True)

# 
print(" :")
print(df.head())
print(df.tail())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# 1 : 10, : 50
df["SMA_10"] = df["Close"].rolling(window=10).mean()
df["SMA_50"] = df["Close"].rolling(window=50).mean()

# 2 
df["Volatility"] = df["Close"].rolling(window=10).std()

# 3 RSI
def calculate_rsi(data, window=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

df["RSI"] = calculate_rsi(df["Close"], window=14)

# 4 
df.dropna(inplace=True)

# 5 
features = ["SMA_10", "SMA_50", "Volatility", "RSI"]
target = "Close"

# 6 80%20%
train_size = 0.8
train, test = train_test_split(df, train_size=train_size, shuffle=False)

X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]

print(" ")

# 
print(f" : {X_train.shape}, : {X_test.shape}")
```

```python
# 1 
model = DecisionTreeRegressor(max_depth=5, random_state=42)

# 2 
model.fit(X_train, y_train)

# 3 
y_pred = model.predict(X_test)

# 4 RMSE: Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f" RMSE: {rmse:.2f}")

# 5 
plt.figure(figsize=(12, 6))
plt.plot(test.index, y_test, label="Actual Price", color="blue", linewidth=2)
plt.plot(test.index, y_pred, label="Predicted Price", color="red", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("BTC/USD Price")
plt.title("BTC/USD Price Prediction with Decision Tree")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# 1 
rf_model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)

# 2 
rf_model.fit(X_train, y_train)

# 3 
y_pred_rf = rf_model.predict(X_test)

# 4 RMSE: Root Mean Squared Error
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print(f" RMSE: {rmse_rf:.2f}")

# 5 
plt.figure(figsize=(12, 6))
plt.plot(test.index, y_test, label="Actual Price", color="blue", linewidth=2)
plt.plot(test.index, y_pred_rf, label="Predicted Price (Random Forest)", color="green", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("BTC/USD Price")
plt.title("BTC/USD Price Prediction with Random Forest")
plt.legend()
plt.show()
```

```python
import xgboost as xgb

# 1 XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# 2 XGBoost
params = {
    "objective": "reg:squarederror",  # 
    "eval_metric": "rmse",  # 
    "eta": 0.05,  # 
    "max_depth": 6,  # 
    "subsample": 0.8,  # 
    "colsample_bytree": 0.8,  # 
}

# 3 XGBoost
num_round = 100
xgb_model = xgb.train(params, dtrain, num_round)

# 4 
y_pred_xgb = xgb_model.predict(dtest)

# 5 RMSE
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
print(f" XGBoostRMSE: {rmse_xgb:.2f}")

# 6 
plt.figure(figsize=(12, 6))
plt.plot(test.index, y_test, label="Actual Price", color="blue", linewidth=2)
plt.plot(test.index, y_pred_xgb, label="Predicted Price (XGBoost)", color="orange", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("BTC/USD Price")
plt.title("BTC/USD Price Prediction with XGBoost")
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

# 1 LSTM
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[["SMA_10", "SMA_50", "Volatility", "RSI", "Close"]])

# 2 30
def create_sequences(data, seq_length=30):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length, :-1])  # 
        y.append(data[i + seq_length, -1])  # 
    return np.array(X), np.array(y)

seq_length = 30  # 30
X, y = create_sequences(scaled_data, seq_length)

# 80%
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f" LSTM: {X_train.shape}, : {X_test.shape}")

# 3 LSTM
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(seq_length, X.shape[2])),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

# 4 
model.compile(optimizer="adam", loss="mean_squared_error")

# 5 
model.fit(X_train, y_train, batch_size=32, epochs=10, validation_data=(X_test, y_test))

# 6 
y_pred_lstm = model.predict(X_test)

# 7 
y_test_original = scaler.inverse_transform(np.concatenate((X_test[:, -1, :], y_test.reshape(-1, 1)), axis=1))[:, -1]
y_pred_lstm_original = scaler.inverse_transform(np.concatenate((X_test[:, -1, :], y_pred_lstm), axis=1))[:, -1]

# 8 RMSE
rmse_lstm = np.sqrt(mean_squared_error(y_test_original, y_pred_lstm_original))
print(f" LSTMRMSE: {rmse_lstm:.2f}")

# 9 
plt.figure(figsize=(12, 6))
plt.plot(test.index[-len(y_test_original):], y_test_original, label="Actual Price", color="blue", linewidth=2)
plt.plot(test.index[-len(y_test_original):], y_pred_lstm_original, label="Predicted Price (LSTM)", color="purple", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("BTC/USD Price")
plt.title("BTC/USD Price Prediction with LSTM")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# 1 StandardScaler 
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[["SMA_10", "SMA_50", "Volatility", "RSI", "Close"]])

# 2 30
def create_sequences(data, seq_length=30):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length, :-1])  # 
        y.append(data[i + seq_length, -1])  # 
    return np.array(X), np.array(y)

seq_length = 30  # 30
X, y = create_sequences(scaled_data, seq_length)

# 80%
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# 3 LSTM
model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(seq_length, X.shape[2])),
    Dropout(0.2),
    LSTM(100, return_sequences=False),
    Dropout(0.2),
    Dense(50),
    Dense(1)
])

# 4 
model.compile(optimizer="adam", loss="mean_squared_error")

# 5 50
model.fit(X_train, y_train, batch_size=32, epochs=50, validation_data=(X_test, y_test))

# 6 
y_pred_lstm = model.predict(X_test)

# 7 
y_test_original = scaler.inverse_transform(np.concatenate((X_test[:, -1, :], y_test.reshape(-1, 1)), axis=1))[:, -1]
y_pred_lstm_original = scaler.inverse_transform(np.concatenate((X_test[:, -1, :], y_pred_lstm), axis=1))[:, -1]

# 8 RMSE
rmse_lstm = np.sqrt(mean_squared_error(y_test_original, y_pred_lstm_original))
print(f" LSTMRMSE: {rmse_lstm:.2f}")

# 9 
plt.figure(figsize=(12, 6))
plt.plot(test.index[-len(y_test_original):], y_test_original, label="Actual Price", color="blue", linewidth=2)
plt.plot(test.index[-len(y_test_original):], y_pred_lstm_original, label="Predicted Price (Improved LSTM)", color="purple", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("BTC/USD Price")
plt.title("BTC/USD Price Prediction with Improved LSTM")
plt.legend()
plt.show()
```

```python
# 1 
initial_balance = 10000  # 
balance = initial_balance
position = 0  # BTC
trade_history = []  # 
trade_threshold = 0.000000025  #  0.25% 

# 
for i in range(len(y_pred_lstm_original) - 1):  #  -1
    actual_price = y_test_original[i]
    predicted_price = y_pred_lstm_original[i + 1]  # 1
    price_change = (predicted_price - actual_price) / actual_price  # 

    # 
    if i % 500 == 0:
        print(f"Time: {test.index[i]}, Actual: {actual_price:.2f}, Predicted: {predicted_price:.2f}, Change: {price_change:.4f}")

    # :  0.25%
    if price_change > trade_threshold and balance >= actual_price:
        position = balance / actual_price  # BTC
        balance = 0  # 
        trade_history.append(("BUY", actual_price, test.index[i]))
        print(f"BUY at {test.index[i]} - Price: {actual_price:.2f}")

    # :  0.25%
    elif price_change < -trade_threshold and position > 0:
        balance = position * actual_price  # BTC
        position = 0  # 
        trade_history.append(("SELL", actual_price, test.index[i]))
        print(f"SELL at {test.index[i]} - Price: {actual_price:.2f}")

# 
final_balance = balance + (position * y_test_original[-1])
profit = final_balance - initial_balance

print(f" : ${initial_balance:.2f}")
print(f" : ${final_balance:.2f}")
print(f" : ${profit:.2f}")

# 
buy_prices = [price for action, price, time in trade_history if action == "BUY"]
sell_prices = [price for action, price, time in trade_history if action == "SELL"]
buy_dates = [time for action, price, time in trade_history if action == "BUY"]
sell_dates = [time for action, price, time in trade_history if action == "SELL"]

plt.figure(figsize=(12, 6))
plt.plot(test.index[-len(y_test_original):], y_test_original, label="Actual Price", color="blue", linewidth=2)
plt.scatter(buy_dates, buy_prices, marker="^", color="green", label="Buy Signal", alpha=1)
plt.scatter(sell_dates, sell_prices, marker="v", color="red", label="Sell Signal", alpha=1)
plt.xlabel("Date")
plt.ylabel("BTC/USD Price")
plt.title("BTC/USD Trading Strategy Backtest (Final Fixed)")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
import matplotlib.font_manager as fm

# 7  'test' 
# X_test 
test_indices = df.index[-len(y_test_original):]
test = df.iloc[-len(y_test_original):].copy()
test = test.set_index(test_indices)

# 1. 
pred_df = pd.DataFrame({
    'Actual': y_test_original,
    'Predicted': y_pred_lstm_original
}, index=test.index)

# 2. 
# 1-1
pred_df['Signal'] = np.where(pred_df['Predicted'] > pred_df['Actual'], 1, -1)

# 3. 
# 'Close' 
test['Return'] = test['Close'].pct_change().shift(-1)  # 

# 4. 
pred_df = pred_df.join(test['Return'])

# 5. NaN
pred_df.dropna(inplace=True)

# 6. 
pred_df['Strategy Return'] = pred_df['Signal'] * pred_df['Return']

# 7. 
pred_df['Cumulative Market Return'] = (1 + pred_df['Return']).cumprod()
pred_df['Cumulative Strategy Return'] = (1 + pred_df['Strategy Return']).cumprod()

# 8. 
# 
final_market_return = pred_df['Cumulative Market Return'].iloc[-1]
final_strategy_return = pred_df['Cumulative Strategy Return'].iloc[-1]

# 
mean_return = pred_df['Strategy Return'].mean()
std_return = pred_df['Strategy Return'].std()
sharpe_ratio = mean_return / std_return * np.sqrt(252)  # 252

print(f" :")
print(f"  : {final_market_return:.2f}")
print(f"  : {final_strategy_return:.2f}")
print(f"  : {sharpe_ratio:.2f}")

# 9. 
font_path = '/System/Library/Fonts/Hiragino Sans GB.ttc'

try:
    font_prop = fm.FontProperties(fname=font_path)
    plt.rcParams['font.family'] = font_prop.get_name()
except FileNotFoundError:
    print("Japanese font is missing")

# 10. 
plt.figure(figsize=(14, 7))
plt.plot(pred_df.index, pred_df['Cumulative Market Return'], label='Buy and Hold', color='blue')
plt.plot(pred_df.index, pred_df['Cumulative Strategy Return'], label='LSTM Strategy', color='purple')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.title(': LSTM vs Buy and Hold')
plt.legend()
plt.grid(True)
plt.show()

# 11. 
# 
pred_df['Position'] = pred_df['Signal'].shift(1)
pred_df['Strategy Return'] = pred_df['Position'] * pred_df['Return']

# 
num_trades = pred_df['Signal'].diff().abs().sum() / 2  # 
print(f" : {int(num_trades)}")

# 12. 
# 
from sklearn.metrics import classification_report, confusion_matrix

# 
# 
print("\n :")
print(pred_df['Strategy Return'].describe())
```
